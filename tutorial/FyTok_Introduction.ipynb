# FyTok 概述

![FyTOK-framework](./figures/FyTok_framework.png)

随着磁约束聚变研究的深入，集成建模和分析程序的开发已经从原本单一、孤立的专有数值计算程序，向开放、灵活的数据处理平台发展，更加强调数据的共享、交流和管理。因此，需要一个集成建模平台，来支持托克马克不同物理模型之间的数据交互、模型之间的调度及物理量的动态关系的表述。

FyTok是自主开发的一套于物理数据分析的集成建模系统。它从物理模块输入、输出数据的标准化、不通语言的程序的标准化封装、到基于物理量关系的动态耦合交互这一系列全生命周期过程集成再统一的平台下。下图给出基本框架，它主要包含下面几个方面：

集成建模系统的几个组成部分：

- 物理模块（physical code/native code）：科研工作者开发的用于解决某个问题的物理程序/模型
- 标准数据交互：不同数据源在标准数据模型下的交互
- actor：actor使用其内部机制（“封装器”）来调用本地的物理代码。物理代码可以是任何语言编写的，是工作流中的一个基本组件，用来执行某些操作，如计算，可视化等。用于actor的API严格依赖于当前的工作流系统。通常，当前actor会使用前一个actor提供的结果，并为下一个actor生成输入数据。在FyTok中，每个物理 IMAS DD为基础构建模块：每个模块代表一个子系统或相对独立概念的完整描述；可显式定义模块之间的依赖关系；**模块和actor的关系是什么？**
- 基于物理关系的工作流：在集成建模系统中，actor的计算是相互依赖的。通常，actor的执行顺序及data的传递方式由所谓的“工作流系统”管理。这种管理可以是简单的脚本（代码），也可以是更复杂的“调度协调器”（如muscle3，kepler等）。工作流中的调度系统会出发actor的特定操作

在FyTok中，将Tokmak作为整体对象来描述。对外提供**FyTok.Tokmak**根节点入口，将实验数据和装置信息整合在一起，来描述不同子系统信息，并进一步和已封装的物理模块联动来跟踪物理量的动态演化。

数据的绑定
- 本地数据的绑定，Python原始数据格式
- 外部不同数据源的绑定
- 不同的数据源在SpDB中被组织成以IDS为单元的树状结构，
- 每个IDS是对装置子系统或者物理的概念的描述 
- 数据来表示装置的状态
- 每个时刻的tokmak装置都包含外部的静态的几何装置描述，如PF线圈信息，TF线圈线圈，WALL的信息等，和内部的动态的物理现象，如平衡，输运，及影响芯部状态的加热、电流驱动等外部作用。
- 当我们把数据按照这样的树状结构整理后，每个静态的数据可以反应某一时刻装置的状态。
- 但是，集成建模的目的是希望能够通过物理程序来反应或者追踪装置内的动态变化的物理过程。那我们就需要利用已知的静态数据来推进或者更新某个时刻或者某些连续时刻装置的状态。
